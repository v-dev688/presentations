```python
import pandas as pd
df = pd.read_csv("GSE132465_GEO_processed_CRC_10X_cell_annotation.txt.gz",
                 sep="\t",  # Assuming tab-separated values (TSV format)
                 compression="gzip")
df.head()
```

```python
df.Cell_type.unique()
```

```python
df.Cell_type.value_counts()
```

# Cell type and their counts

```python
df.groupby(["Cell_type","Class"]).size().unstack(fill_value=0)
```

# In epithelial cells

```python
epithelial_df=df[df['Cell_type']=="Epithelial cells"]
epithelial_df.shape
```

```python
epithelial_df.groupby(["Cell_subtype","Class"]).size().unstack(fill_value=0)
```

# In Tcells

```python
Tcell_df=df[df['Cell_type']=="T cells"]
Tcell_df.shape
```

```python
Tcell_df.groupby(["Cell_subtype","Class"]).size().unstack(fill_value=0)
```
